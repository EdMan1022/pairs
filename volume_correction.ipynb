```python
data_set = DataSet()
```

```python
data_set.data.index = pd.to_datetime(data_set.data.index)
data_set.volume.index = pd.to_datetime(data_set.volume.index)
```

```python
dollar_volume = data_set.volume * data_set.data
```

```python
dollar_volume[dollar_volume.mean(axis=1) >= 1.0e8]
```

```python
closed_days = dollar_volume.loc[:, 'AAPL'][dollar_volume.loc[:, 'AAPL'] == 0].index
```

```python
open_index = dollar_volume.index[~dollar_volume.index.isin(closed_days)]
```

```python
dollar_volume.loc[closed_days, :].max(axis=1)
```

```python
corrected_volume = dollar_volume.loc[open_index]
```

```python
corrected_volume.drop('DRYS', axis=1, inplace=True)
```

```python
top_companies = corrected_volume.loc[:, corrected_volume.mean(axis=0) >= 1.0e8]
```

```python
imputer = Imputer().fit(X=top_companies)
```

```python
top_companies = pd.DataFrame(imputer.transform(top_companies), index=top_companies.index, columns=top_companies.columns)
```

```python
scaler = StandardScaler().fit(top_companies)
```

```python
scaled_companies = pd.DataFrame(scaler.transform(top_companies), index=top_companies.index, columns=top_companies.columns)
```

```python
scaled_corr = scaled_companies.corr()
original_corr = top_companies.corr()
```
